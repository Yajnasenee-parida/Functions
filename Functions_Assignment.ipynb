1.  In Python, both functions and methods are blocks of reusable code, but they differ in their context and usage:

- **Function:** A function is a standalone block of code that is not tied to any object. It can be called independently and does not require an object to operate. For example:

def greet(name):

     return f"Hello, {name}!"

print(greet("Alice")) # Output: Hello, Alice!

- **Method:** A method is a function that is associated with an object. It is defined within a class and operates on the data contained within that class. Methods are called on objects using the dot notation. For example:

class Greeter:

      def greet(self, name):

           return f"Hello, {name}!"

greeter = Greeter()

print(greeter.greet("Alice")) # Output: Hello, Alice!

5.  In Python, iterables and iterators are fundamental concepts that facilitate the process of looping over data structures. Understanding the difference between them is crucial for efficient data processing.

**Iterable**

An iterable is any Python object capable of returning its members one at a time, allowing it to be looped over in a for-loop. Common examples include lists, tuples, dictionaries, and sets. These objects implement the __iter__() method, which returns an iterator.

**Example:**

 #List is an iterable

my_list = [1, 2, 3]

for item in my_list:

     print(item)

**Iterator**

An iterator is an object that represents a stream of data. It implements two methods: __iter__() and __next__(). The __iter__() method returns the iterator object itself, and the __next__() method returns the next item from the stream. When no more data is available, __next__() raises a StopIteration exception.

**Example:**

 #Creating an iterator from a list

my_list = [1, 2, 3]

my_iterator = iter(my_list)

print(next(my_iterator)) # Output: 1

print(next(my_iterator)) # Output: 2

print(next(my_iterator)) # Output: 3

**Practical Example**

Consider a scenario where you want to iterate over a list of cities:

 #List of cities (iterable)

cities = ["Berlin", "Vienna", "Zurich"]

 #Creating an iterator from the list

iterator_obj = iter(cities)

 #Iterating using the iterator

print(next(iterator_obj)) # Output: Berlin

print(next(iterator_obj)) # Output: Vienna

print(next(iterator_obj)) # Output: Zurich


6.  A generator object in Python is a special type of iterator created by a generator function or a generator expression. Unlike regular functions that return a single value and terminate, generator functions use the yield keyword to produce a sequence of values lazily, pausing their execution and maintaining their state between iterations.

- **Lazy Evaluation:** Generators compute values on demand, making them memory-efficient, especially for large datasets or infinite sequences.

- **State Retention:** After yielding a value, the generator pauses and retains its state, allowing it to resume execution from where it left off.

- **Iterator Protocol:** Generator objects implement the iterator protocol, meaning they have a __next__() method to fetch the next value and raise StopIteration when exhausted.

**Example of a Generator Function**

Hereâ€™s a simple example of a generator function that yields numbers from 1 to a given maximum:

def number_generator(max_num):

       num = 1
       while num <= max_num:
           yield num
           num += 1
 #Create a generator object

gen = number_generator(5)

 #Iterate through the generator

for value in gen:

     print(value)

**output**

1

2

3

4

5

In this example, the yield statement pauses the function, returning the current value of num. When next() is called (implicitly by the for loop), the function resumes from the last yield.


7.  Generators offer significant advantages over regular functions, including memory efficiency, lazy evaluation, and ease of creating iterators, making them ideal for handling large datasets and infinite sequences.

- **Memory Efficiency:**

Generators produce values one at a time and do not store the entire sequence in memory. This is particularly beneficial when working with large datasets or streams of data, as it reduces memory consumption significantly compared to regular functions that return complete lists or collections.

- **Lazy Evaluation:**

Generators use lazy evaluation, meaning they compute values only when requested. This allows for immediate use of the first few items while the rest are generated on-the-fly, which can lead to performance improvements in scenarios where not all data is needed at once.

- **Simplified Iterator Creation:**

Generators simplify the process of creating iterators. Instead of implementing the __iter__ and __next__ methods, you can simply use the yield statement in a function. This makes the code cleaner and easier to understand.

- **Handling Infinite Sequences:**

Generators can represent infinite sequences without running out of memory. For example, a generator can produce an infinite series of numbers, yielding one at a time as needed, which is impossible with regular functions that return a complete list.

- **Improved Performance in Certain Scenarios:**

In cases where you only need a subset of results or when the computation of results is expensive, generators can save time and resources by avoiding unnecessary calculations. They allow you to start processing data immediately while the rest is still being generated.


8.  A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike regular functions created with the def keyword, lambda functions are concise and typically used for short, immediate operations. They consist of a single expression and can take any number of arguments.

10.  Python's map(), filter(), and reduce() functions are powerful tools for functional programming, enabling concise and efficient operations on iterables. Here's an overview of each function with examples:

**map() Function**

The map() function applies a given function to each item in an iterable (like a list or tuple) and returns a map object (an iterator). It is commonly used for transforming data.

**Syntax:** map(function, iterable)

**Example:**

 #Double each number in a list

def double(n):

     return n * 2

numbers = [1, 2, 3, 4]

result = map(double, numbers)

print(list(result)) # Output: [2, 4, 6, 8]


**filter() Function**

The filter() function filters elements from an iterable based on a function that returns True or False. It returns an iterator containing only the elements that satisfy the condition.

**Syntax:** filter(function, iterable)

**Example:**

 #Filter even numbers from a list

def is_even(n):

     return n % 2 == 0

numbers = [1, 2, 3, 4, 5, 6]

even_numbers = filter(is_even, numbers)

print(list(even_numbers)) # Output: [2, 4, 6]

**reduce() Function**

The reduce() function applies a function cumulatively to the items of an iterable, reducing it to a single value. Unlike map() and filter(), it is not built-in and must be imported from the functools module.

**Syntax:** reduce(function, iterable[, initial])

**Example:**

from functools import reduce

 #Compute the product of a list

numbers = [1, 2, 3, 4]

product = reduce(lambda x, y: x * y, numbers)

print(product) # Output: 24

```python
# 5
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # Total number of terms to generate
        self.current = 0       # Current term index
        self.a, self.b = 0, 1  # Initial two Fibonacci numbers

    def __iter__(self):
        return self  # An iterator must return itself

    def __next__(self):
        if self.current >= self.n_terms:
            raise StopIteration  # Stop when the specified number of terms is reached
        if self.current == 0:
            self.current += 1
            return self.a  # Return the first Fibonacci number
        elif self.current == 1:
            self.current += 1
            return self.b  # Return the second Fibonacci number
        else:
            self.a, self.b = self.b, self.a + self.b  # Update Fibonacci numbers
            self.current += 1
            return self.a  # Return the next Fibonacci number

# Example usage:
fib = FibonacciIterator(10)  # Generate the first 10 Fibonacci numbers
for num in fib:
    print(num)

```

```python
# 6
def powers_of_two(max_exponent):
    """Generator that yields powers of 2 up to the given exponent."""
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

# Example usage:
for power in powers_of_two(5):
    print(power)
```

```python
# 7
def read_file_line_by_line(file_path):
    """Generator function to read a file line by line."""
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()  # Yield each line without leading/trailing whitespace
    except FileNotFoundError:
        print(f"Error: The file at '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
file_path = 'example.txt'

for line in read_file_line_by_line(file_path):
    print(line)
```

```python
# 9
# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Using map() to apply the conversion function to each element in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print("Temperatures in Fahrenheit:", fahrenheit_temperatures)
```

```python
# 10
def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    # Use filter to exclude vowels
    result = ''.join(filter(lambda char: char not in vowels, input_string))
    return result

# Example usage
input_string = "Hello, World!"
output_string = remove_vowels(input_string)
print("Original String:", input_string)
print("String without vowels:", output_string)
```

```python
# 11
# Sample data: [order_number, (quantity, price_per_item)]
orders = [
    [1, (4, 40.95)],
    [2, (5, 56.80)],
    [3, (3, 32.95)],
    [4, (3, 24.99)]
]

# Function to calculate the total amount for each order
calculate_order = lambda order: (
    order[0],  # Order number
    order[1][0] * order[1][1] + (10 if order[1][0] * order[1][1] < 100 else 0)  # Total with adjustment
)

# Use map to apply the function to all orders
result = list(map(calculate_order, orders))

# Output the result
print(result)
```
